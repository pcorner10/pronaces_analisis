```python
columns_to_check = [                                'Clave del Proyecto:',
                    'Localidad, Mpio. y Edo.:',                         'Fecha:',
                    'Tipo de muestreo:',       'Hora de inicio (hh:mm): ',
                     'Tipo de muestra:',     'Hora de término (hh:mm):',
                     'UTM hor.:','UTM ver.:',

                   'Refrigeración:',        'Condiciones climáticas:',
                        'Uso de suelo:',            'Tipo de vegetación:',
       'Edificación o infraestructura:',             'Forma del terreno:',
                         'Inclinación:',          'Evidencia de erosión:',
                             'Paisaje:',             'Material parental:',
                      'Toma de muestra',              'Revisó y Autorizó',
                             'Nombre: ',                         'Firma:']

columns_to_check = [
       'Clave del Proyecto:', 'Localidad, Mpio. y Edo.:','Nombre de la(s) persona(s) que participan:',
       'Fecha:', 

]
```

```python
def readFormato(fileExcel, full_df):
  """_summary_: Lee el formato de la base de datos de suelos alojados en hojas dentro de un archivo excel y lo retorna en un dataframe
  """
  # load excel file and get sheet names
  xls = pd.ExcelFile(fileExcel)
  sheet_names = xls.sheet_names
  
  #full_df = pd.DataFrame()
  
  for sheet in sheet_names:
    df = pd.read_excel(fileExcel, sheet_name=sheet)
    df.dropna(axis=0, how='all', inplace=True)
    
    # drop columns with all NaN values
    df.dropna(axis=1, how='all', inplace=True)
    
    
    diccionario_dataframe = {}

    for i, row in df.iterrows():
      
      # iterate over elements in row
      for j, element in enumerate(row):
        # check if element is in columns_to_check
        if element in columns_to_check:
          # get column name
          col_name = element
          # get value in the next column two columns
          val = row[j+1:j+4]
          # drop NaN values
          val = val.dropna()
          
          if len(val) != 0:
            diccionario_dataframe[col_name] = val.values[0]
    full_df = pd.concat([full_df, pd.DataFrame(diccionario_dataframe, index=[0])], ignore_index=True)
  return full_df, df
  
```
