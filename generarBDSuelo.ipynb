# Proceso para generar la base de datos de suelo

## Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob, os
```

## files

```python
files_suelos = glob.glob("data/suelo/raw/*.xlsx")
files_suelos
```

```python
columns_to_right = [
       'Clave del Proyecto:',
       'Localidad, Mpio. y Edo.:',
       
       'Fecha:',
       'Hora de inicio (hh:mm): ',
       'Hora de término (hh:mm):',
       
       'Tipo de muestreo:',
       'Tipo de muestra:',
       'Refrigeración:    ',
       
       'UTM hor.:',
       'UTM ver.:',
       'msnm:',
       
       'Condiciones climáticas:',
       'Tipo de vegetación:',
       'Forma del terreno:',
       'Tipo de suelo:',
       'Paisaje:',
       
       'Uso de suelo:',
       'Edificación o infraestructura:',
       'Inclinación:',
       'Evidencia de erosión:',
       'Material parental:',
       ]

columns_to_down = [
       'Clave de la muestra:'
]
```

```python
def readFormato(fileExcel, full_df):
  """_summary_: Lee el formato de la base de datos de suelos alojados en hojas dentro de un archivo excel y lo retorna en un dataframe
  """
  # load excel file and get sheet names
  xls = pd.ExcelFile(fileExcel)
  sheet_names = xls.sheet_names
  
  #full_df = pd.DataFrame()
  
  for sheet in sheet_names:
    df = pd.read_excel(fileExcel, sheet_name=sheet)
    df.dropna(axis=0, how='all', inplace=True)
    
    # drop columns with all NaN values
    df.dropna(axis=1, how='all', inplace=True)
    
    
    diccionario_dataframe = {}
    ubicacion = 0
    contador = 0
    for i, row in df.iterrows():
      

      # iterate over elements in row
      for j, element in enumerate(row):
        if element == 'Clave de la muestra:':
          ubicacion = j
          contador = 1
          continue
        
        if element == 'Refrigeración:    ':
          val = row[j+1:j+4]
          # drop NaN values
          val = val.dropna()
          
          if val.values[0] == ' Sí__X___            No_____':
            
            diccionario_dataframe['Refrigeración:'] = 'No'
          else:
            diccionario_dataframe['Refrigeración:'] = 'Sí'
        
        if j == ubicacion and contador == 1: 
          diccionario_dataframe['Clave de la muestra:'] = element
          contador = 2
        
        # check if element is in columns_to_check
        if element in columns_to_right:
          # get column name
          col_name = element
          # get value in the next column two columns
          val = row[j+1:j+4]
          # drop NaN values
          val = val.dropna()
          
          # check if len of val different than 0 and not in columns_to_right
          if len(val) != 0 and val[0] not in columns_to_right:
            diccionario_dataframe[col_name] = val.values[0]
    full_df = pd.concat([full_df, pd.DataFrame(diccionario_dataframe, index=[0])], ignore_index=True)
  return full_df, df
  
```

```python
dfFull = pd.DataFrame()
for file in files_suelos:
  dfFull, df_suplementario = readFormato(file, dfFull)
  
dfFull
```
